```python
import random
import math

def strength(value):
    x= value
    return math.log2(x+1) + (x/10)

def utility(maxV, minV):
    i = random.randint(0, 1)
    rendom = (random.randint(1,10))/10
    return strength(maxV) - strength(minV) + ((-1)**i)*rendom

def maxval(depth, alpha, beta, maxV, minV):
    if depth == 5:
        return utility(maxV, minV)
    v = -float('inf')
    for branch in range(2):
        v = max(v, minval(depth + 1, alpha, beta, maxV, minV))
        if v >= beta:
            return v
        alpha = max(alpha, v)
    return v

def minval(depth, alpha, beta, maxV, minV):
    if depth == 5:
        return utility(maxV, minV)
    v = float('inf')
    for branch in range(2):
        v = min(v, maxval(depth + 1, alpha, beta, maxV, minV))
        if v <= alpha:
            return v
        beta = min(beta, v)
    return v


def alphabeta(maxV, minV, csmax):
    if csmax:
        return maxval(0, -float('inf'), float('inf'), maxV, minV)
    else:
        return minval(0, -float('inf'), float('inf'), maxV, minV)

def gameplay(start_player, cs_strength, ca_strength):
    cs_score = 0
    ca_score = 0
    draws = 0

    if start_player == 0:
        csismax = [True, False, True, False]
    else:
        csismax = [False, True, False, True]

    for round in range(4):
        if csismax[round]:
            maxV = cs_strength
            minV = ca_strength
        else:
            maxV = ca_strength
            minV = cs_strength
        
        result = alphabeta(maxV, minV, csismax[round])
        if result>0:
            winner = "Magnus Carlsen"
        elif result<0:
            winner = "Fabiano Caruana"
        else:
            winner = "Draw"

        if csismax[round]:
            if winner == "Magnus Carlsen":
                winner_role = "Max"
            else:
                winner_role = "Min"
        else:
            if winner != "Magnus Carlsen":
                winner_role = "Max"
            else:
                winner_role = "Min"

        if winner == "Magnus Carlsen":
            cs_score += 1
        elif winner == "Fabiano Caruana":
            ca_score += 1
        else:
            draws += 1
        
        print(f"Game {round + 1} Winner: {winner} ({winner_role}) (Utility value: {result:.2f})")

    print("\nOverall Results:")
    print(f"Magnus Carlsen Wins: {cs_score}")
    print(f"Fabiano Caruana Wins: {ca_score}")
    print(f"Draws: {draws}")
    
    if cs_score > ca_score:
        print("Overall Winner: Magnus Carlsen")
    elif ca_score > cs_score:
        print("Overall Winner: Fabiano Caruana")
    else:
        print("Overall Winner: Draw")
        


def chesh():
    start_player = int(input("Enter starting player for game 1 (0 for Carlsen, 1 for Caruana): "))
    cs_strength = float(input("Enter base strength for Carlsen: "))
    ca_strength = float(input("Enter base strength for Caruana: "))
    
    gameplay(start_player, cs_strength, ca_strength)

chesh()
```
